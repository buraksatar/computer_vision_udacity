# Hough Circle Detection

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
# Read in the image
image = cv2.imread('images/round_farms.jpg')

# Change color to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Gray and blur
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

gray_blur = cv2.GaussianBlur(gray, (3,3), 0)

plt.imshow(gray_blur, cmap='gray')
```

## HoughCircles function

HoughCircles taes in a few things as its arguments:
- an input image, detection method (Hough gradient), resolution factor between the detection and image,
- minDist - the minimum distance between circles
- param1 - the higher value for performing Canny edge detection
- param2 - threshold for cicle detectin, a smaller value --> more circles will be detected
- min/max radius for detecte circles

The variable we should change will be the last two: min/max radius for detected circles.

```python
##### for drawing circles on
circles_im = np.copy(image)

# use HoughCircles to detect circles

circles = cv2.HoughCircles(gray_blur, 
                           cv2.HOUGH_GRADIENT, 1,
                           minDist=50,
                           param1=70,
                           param2=11,
                           minRadius=20,
                           maxRadius=35)

# convert circles into expected type
circles = np.uint16(np.around(circles))

# draw each one
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(circles_im,
               (i[0],i[1]),i[2],
               (0,255,0),2)
    # draw the center of the circle
    cv2.circle(circles_im,
               (i[0],i[1]),2,
               (0,255,0),3)
    
plt.imshow(circles_im)
print('Circles shape: ', circles.shape)
```
